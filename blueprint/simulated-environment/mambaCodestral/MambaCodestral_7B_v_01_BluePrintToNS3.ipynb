# Generate the NS3 code for the smart_agriculture scenario.

```python
with open("sample_blueprint.json") as f:
        sample_blueprint = f.read()
with open("sample_ns3_code.cc") as f:
        sample_ns3_code = f.read()
with open("smart_agriculture_complete.json") as f:
        test_blueprint = f.read()
```

```python
prompt = f"""
        Given the following TEST BLUEPRINT provide me the corresponding NS-3 code. Use the follwing SAMPLE NS-3 code and SAMPLE BLUEPRINT as reference.
        Not add description, just return cpp code.
                TEST BLUEPRINT:

                {test_blueprint}

                SAMPLE BLUEPRINT:

                {sample_blueprint}

                SAMPLE NS3 CODE:

                {sample_ns3_code}


                NS3 CODE:

                """
```

```python
message = [
    {
        "role": "user",
        "content": prompt
    }
]

chat_response = client.chat.complete(
    model=model,
    messages=message
)
print(chat_response.choices[0].message.content)
```

## Save the .cc code into a file.

```python
with open("smart_agriculture.cc", "w") as file:
    file.write(chat_response.choices[0].message.content)
```
