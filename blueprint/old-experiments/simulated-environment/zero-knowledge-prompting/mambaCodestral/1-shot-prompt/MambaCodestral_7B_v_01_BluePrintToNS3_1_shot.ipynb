# Generate the NS3 code for the smart_agriculture scenario.

```python
with open("sample_blueprint.json") as f:
        sample_blueprint = f.read()
with open("sample_ns3_code.cc") as f:
        sample_ns3_code = f.read()
with open("smart_agriculture.json") as f:
        test_blueprint = f.read()
```

```python
prompt = """
 Given the following SAMPLE BLUEPRINT and the corresponding SAMPLE NS3 code, use them as a sample to generate the NS3 code for the TEST BLUEPRINT:

        SAMPLE BLUEPRINT:

        """+sample_blueprint+ """

        SAMPLE NS3 CODE:

        """+sample_ns3_code+"""

        TEST BLUEPRINT:

        """+test_blueprint+"""

        TEST NS3 CODE:

        """
print(prompt)
```

```python
message = [
    {
        "role": "user",
        "content": prompt
    }
]

chat_response = client.chat.complete(
    model=model,
    messages=message
)
print(chat_response.choices[0].message.content)
```

## Save the .cc code into a file.

```python
with open("smart_agriculture-1_shot.cc", "w") as file:
    file.write(chat_response.choices[0].message.content)
```
